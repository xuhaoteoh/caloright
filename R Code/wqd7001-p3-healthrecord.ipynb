```r


# https://www.calculator.net/calorie-calculator.html

library(shiny)

# Define UI for slider demo app ----
ui <- fluidPage(
  
  # App title ----
  titlePanel("Health Record"),
  
  # Sidebar layout with input and output definitions ----
  sidebarLayout(
    
    # Sidebar to demonstrate various slider options ----
    sidebarPanel(
      
      sliderInput("age", "Age:",
                  min = 0, max = 120,
                  value = 25),
      
      selectInput(inputId = "gender",
                  label = "Gender:",
                  choices = c("male", "female")),
      
      sliderInput("height", "Height(cm):",
                  min = 0, max = 250,
                  value = 180),

      sliderInput("weight", "Weight(kg):",
                  min = 0, max = 250,
                  value = 65),
      
      selectInput(inputId = "activity",
                  label = "Activity:",
                  choices = c('Basal Metabolic Rate (BMR)','Sedentary: little or no exercise','Light: exercise 1-3 times/week','Moderate: exercise 4-5 times/week','Active: daily exercise or intense exercise 3-4 times/week',"Very Active: intense exercise 6-7 times/week", "Extra Active: very intense exercise daily, or physical job")),
           
    ),
    
    # Main panel for displaying outputs ----
    mainPanel(
      
      # Output: Table summarizing the values entered ----
      tableOutput("values")
      
    )
  )
)

# Define server logic for slider examples ----
server <- function(input, output) {
  
  # Reactive expression to create data frame of all input values ----
  sliderValues <- reactive({
    
    data.frame(
      Name = c("Age",
               "Gender",
               "Height",
               "Weight",
               "Activity"),
      Value = as.character(c(input$age,
                             input$gender,
                             input$height,
                             input$weight,
                             input$activity)),
      stringsAsFactors = FALSE)
  })
  
  # Show the values in an HTML table ----
  output$values <- renderTable({
    sliderValues()
  })
  
}

# Create Shiny app ----
shinyApp(ui, server)
```
