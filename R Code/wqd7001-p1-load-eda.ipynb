```r
# 2.1 Univariate Column Statistics

summary(nutritions)
######### psych package ############
library(psych)
psych::describe(nutritions)
#### dataMaid package ##########
library(dataMaid)
summarize(nutritions)
makeDataReport(nutritions, replace=TRUE,output="gg.html")
##### Additional univariate description functions ##########
library(summarytools)
dfSummary(nutritions)

library(inspectdf)
inspect_num(nutritions)
inspect_cat(nutritions)

library(prettyR)
prettyR::describe(nutritions)

library(Hmisc)
Hmisc::describe(nutritions)

library(funModeling)
profiling_num(nutritions)

library(skimr)
skim(nutritions)
###Univariate Group Statistics###############
#Use tapply() to group by a factor ############
tapply(nutritions$calories,nutritions$fat,mean,na.rm=TRUE)
#Within groups, print detailed statistical information of the var under examination.
library(psych)
describeBy(nutritions$fat,group=nutritions$calories)

###Use dplyr functions #################

###Plots of Univariate Statistics (boxplot,histogram)####
ggplot(nutritions,aes(calories,fat))+geom_boxplot()
```

```r
#######OUTLIERS########################################
#Univariate Outlier Detection ##############
###### Interactive histogramM#################
library(plotly)
p1<-ggplot(nutritions,aes(calories,fat))+geom_boxplot()
p1
pp<-ggplotly(p1)
api_create(pp, filename = "boxplot-salary-sex")
##########################################

##############Package extremevalues################################
library(extremevalues)
L<-getOutliers(nutritions$calories)
outlierPlot(nutritions$calories,L)

#### Package OutlierDetection ############
library(OutlierDetection)
UnivariateOutlierDetection(nutritions$calories)
###########################################################
########### Multivariate Outlier Detection ################

####### Package OutlierDetection ##########
library(OutlierDetection)
sal1<-select(nutritions,calories,fat)
saldata<-na.omit(sal1)
salr <- saldata[,c("fat", "calories")]
OutlierDetection(salr)
############ Package Chemometrics###########
library(chemometrics)
gg<-Moutlier(saldata)
mahal_dist<-gg$md
names(mahal_dist) <- 1:nrow(saldata)
ff<-sort(mahal_dist, decreasing = TRUE)
ff
#############Package Mvoutlier###################
library(mvoutlier)
par("mar")
par(mar=c(1,1,1,1))
out <- chisq.plot(saldata)
out$outliers


########AUTOMATIC EDA######################

library(DataExplorer)
create_report(sal,y="calories")

library(efficient)
library(AEDA)
fastReport(data=nutritions)
rmarkdown::render("MainReport.rmd")

library(ExPanDaR)
ExPanD(nutritions)

library(RtutoR)
gg=generate_exploratory_analysis_ppt(sal,target_var="calories",
output_file_name="Rtutorreport.pptx")
```
