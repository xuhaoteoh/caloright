```r
# load nutrition data
nutritions <- read.csv("../input/nutrition-clean/nutrition_cleaned.csv")
```

```r
# SORT
# parameters
sort_index = 4
sort_col_name = 'calories'
data_sorted <- nutritions

# Sort by vector name [z]
data_sorted[
  with(data_sorted, order(sort_col_name)),
]


# Sorting by Column Index
data_sorted[
  order( data_sorted[,sort_index] ),
]
```

```r
# Search
# 1. Searching columns by name
cols_selected = c('name','calories','total_fat')
nutritions[,cols_selected]
```

```r
# 2. Searching by conditions
# By name
searchName <- function(name){
which(nutritions$name==name)
}
nutritions[searchName(name='Cornstarch'), ]
# By calories
searchCalory <- function(calory){
which(nutritions$calories>=calory)
}
nutritions[searchCalory (calory=200), ]
# By total_fat
searchFat <- function(fat){
which(nutritions$total_fat>=fat)
}
nutritions[searchFat(fat=5), ]
# By cholesterol
searchCholesterol <- function(cholesterol){
which(nutritions$cholesterol==cholesterol)
}
nutritions[searchCholesterol(cholesterol='Cornstarch'), ]
# ...
```
